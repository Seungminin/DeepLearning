```python
#red wine과 white wine의 정보를 학습하여 레드 와인인 지 화이트 와인인 지 맞추는 Program을 짜보자!. 지도학습 이항분류 문제.
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint, EarlyStopping

import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import os
```

```python
#seed값 설정
seed = 0
np.random.rand(seed)
tf.random.set_seed(3)

#데이터 입력
df_pre = pd.read_csv('dataset/wine.csv', header = None, skiprows = 1)
df = df_pre.sample(frac=1) #랜덤으로 데이터를 불러온다. frac = 1은 데이터의 100% 데이터를 랜덤. 0.5는 50%만 랜덤

dataset = df.values #pandas 데이터를 이용할 때는 values를 사용하여 dataset에 저장
X = dataset[1:,0:12]
Y = dataset[1:, 12]

#모델 설정
model = Sequential()
model.add(Dense(30, input_dim = 12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

#model 컴파일
model.compile(loss = 'binary_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy']) #정확도를 기록하기 위한 방법

#모델이 저장될 폴더 설정
MODEL_DIR = 'models/'
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)

#모델 저장 조건 설정
modelpath = "models/{epoch:02d}-{val_loss:.4f}.keras"
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', verbose = 1, save_best_only=True)
#verbose = 1 해당 함수의 진행상황 출력, 0이면 출력 X, save_best_only : monitor가 테스트셋 오차니까, 이전 보다 나은 결과가 나오면 모델 저장.

#학습 자동 중단 설정 : EarlyStopping
early_stopping_callback = EarlyStopping(monitor = 'val_loss', patience = 100) #100번까지 기다리고 나아지지 않으면 자동 종료

model.fit(X,Y, validation_split = 0.2, epochs = 3500, batch_size = 500, verbose = 0, callbacks = [early_stopping_callback,checkpointer])
"""
모델 matplotilb.pyplot으로 시각화 하기.
model 실행
history = model.fit(X, Y, validation_split=0.2, epochs = 200, batch_size = 200, verbose = 0, callbacks = [checkpointer])

y_vloss = history.history['val_loss']
y_acc = history.history['accuracy']

x값을 지정하고, 정확도를 파란색, vloss값을 빨간색으로 지정.
x_len = np.arange(len(y_acc))
plt.plot(x_len, y_vloss, 'o', c = 'red', markersize = 3)
plt.plot(x_len, y_acc, 'o', c = 'blue', markersize = 3)

plt.show()
결과 출력
"""
#print("\n accuracy = %.4f" %model.evaluate(X,Y)[1])

```
