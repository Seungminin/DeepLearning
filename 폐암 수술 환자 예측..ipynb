```python
# -*- coding: utf-8 -*-
# 폐암 수술 환자의 생존율 예측하는 Code

# 코드 내부에 한글을 사용가능 하게 해주는 부분입니다.
# 딥러닝을 구동하는 데 필요한 케라스 함수를 불러옵니다.
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 필요한 라이브러리를 불러옵니다.
import numpy as np
import tensorflow as tf

# 실행할 때마다 같은 결과를 출력하기 위해 설정하는 부분입니다.
np.random.seed(3)
tf.random.set_seed(3)

# 준비된 수술 환자 데이터를 불러들입니다.
Data_set = np.loadtxt("dataset/ThoraricSurgery.csv", delimiter=",")

# 환자의 기록과 수술 결과를 X와 Y로 구분하여 저장합니다.
X = Data_set[:,0:17] #Data에서 환자들의 attribute를 나타내는 정보
Y = Data_set[:,17] #수술의 결과를 나타내는 Class (0,1)

# 딥러닝 구조를 결정합니다(모델을 설정하고 실행하는 부분입니다).
model = Sequential()
model.add(Dense(30, input_dim=17, activation='relu')) #은닉층 1 input data를 17개 받아와서, 은닉층 노드를 30개 만든다.
model.add(Dense(1, activation='sigmoid')) #출력층이 선형문제이기 때문에 1이다.

# 딥러닝을 실행합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) #오차 함수 계산법은 binary엔트로피 사용, 최적화는 adam
model.fit(X, Y, epochs=100, batch_size=5)
#ephochs은 모든 샘플데이터를 사용하라이다. ex) epochs = 1 모든 데이터를 한번 사용해라
#batch_size = 데이터를 모델에 넣을 때 몇개 씩 끊어서 넣어라. batch_size는 5이면 5개씩 끊어서 넣어라. 5*94 = 470
```
