```python
df = pd.read_csv('dataset/sonar.csv')
print(df.info()) #0 ~ 59까지 60개의 feature, 60번째는 Class이다.
```

```python
print(df.head())
```

```python
#데이터의 상태를 pandas로 파악을 했다.
from keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

import numpy as np
import tensorflow as tf

#seed값 설정
np.random.seed(3)
tf.random.set_seed(3)

#데이터 불러오기
dataset = df.values
X = dataset[:,0:60]
Y_obj = dataset[:,60]

#문자열 변환
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)

#모델 설정
model = Sequential()
model.add(Dense(24,input_dim = 60, activation = 'relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation = 'sigmoid'))

#model 컴파일
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

#model 실행
model.fit(X,Y,epochs = 200, batch_size = 5)

#결과 출력
print("\n Accuracy = %.4f" %(model.evaluate(X,Y)[1]))
```

과연 Accuracy가 1, 즉 정확도가 100%일까? 아니다!
너무 dataset에만 치중되어 있는 모습이다. 새로운 데이터가 입력이 되면 제대로 된 학습을 못할 것이다.

```python
#따라서 우리는 testset을 별도로 만들어줘서 Overfitting을 방지해야한다.
from sklearn.model_selection import train_test_split
from keras.models import load_model
#학습셋과 test셋의 구분
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state = 3)

model=Sequential()
model.add(Dense(24, input_dim = 60, activation='relu'))
model.add(Dense(10,activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))

model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])

model.fit(X_train, Y_train, epochs = 130, batch_size = 5)
#Model을 저장하는 법
#model.save('my_model.h5')

#del model
#model = load_model('my_model.h5')

#테스트셋에 모델 적용
print("\n Test Accuracy : %0.4f" %(model.evaluate(X_test, Y_test)[1]))
```

```python
#test셋에 데이터의 한계를 보완하고자 K-fold cross validation 방법을 사용하여 테스트에 대한 정확성을 높인다. 
#k-fold cross validation은 만약 100에폭을 한다면 100*k에폭을 하는 것과 같다.
from sklearn.model_selection import StratifiedKFold

#dataset을 10개로 쪼갠다
n_fold = 10
skf = StratifiedKFold(n_splits = n_fold, shuffle = True, random_state = 3)

accuracy = []

for train,test in skf.split(X,Y):
    model = Sequential()
    model.add(Dense(24,input_dim = 60, activation = 'relu'))
    model.add(Dense(10,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    
    model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics=['accuracy'])
    model.fit(X[train], Y[train], epochs = 100, batch_size = 5)
    k_accuracy = "%.4f" %(model.evaluate(X[test],Y[test])[1])
    accuracy.append(k_accuracy)
print("\n %.f fold accuracy :" %n_fold, accuracy)
```
